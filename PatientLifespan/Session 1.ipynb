```python
import pandas as pd
import os 
```

Получаем список всех файлов

```python
DIR = "C:\\Users\\User\\Desktop\\BRICKS\\rest"
files = os.listdir(DIR)
```

Для начала необходимо получить список всх сенсоров сенсоров, тк в дольнейшем с их помощб будем создавать поля для датасета

```python
sensors = set()
for i in files[:30]:
    df = pd.read_csv(DIR+'\\'+i,sep=",",skiprows=1, names=['Time', 'Parameter', 'Value'])
    sensors.update(df.Parameter.values)
sensors = list(sensors)
```

Для всех значений сенсоров одного человека получим начальное значение, конечное, минимальное, максимальное, среднее всех показателей и медиану всех показателей. Также к финальным полям добавим id человека

```python
cols = ["ID"]
for i in sensors:
    for j in ["_first_val", "_last_val", "_min", "_max", "_mean", "_median"]:
        cols.append(i+j)
```

```python
cols
```

Создадим датафрейм с этими полями. Далее будем брать каждый файл, проходить по списку сенсоров, считать необходимые нам значения и заполнять словарь. После этого каждая строчка в датафрейме окажется показателями одного человека.

```python
DF = pd.DataFrame(columns=cols)
```

```python
%%time
for i in files:
    tmpDict = {}
    df = pd.read_csv(DIR+'\\'+i,sep=",",skiprows=2, names=['Time', 'Parameter', 'Value'])
    tmpDict["ID"] = int(i[:i.find(".")])
    for sense in sensors:
        tmpDF = df[df["Parameter"] == sense]
        try:
            tmpDict[sense+"_first_val"] = tmpDF[tmpDF["Parameter"] == sense]["Value"].head(1).values[0]
        except Exception:
            print(end="")                                    
        try:
            tmpDict[sense+"_last_val"] = tmpDF[tmpDF["Parameter"] == sense]["Value"].tail(1).values[0]
        except Exception:
            print(end="")
        try:
            tmpDict[sense+"_min"] = tmpDF[tmpDF["Parameter"] == sense]["Value"].min()
        except Exception:
            print(end="")
        try:
            tmpDict[sense+"_max"] = tmpDF[tmpDF["Parameter"] == sense]["Value"].max()
        except Exception:
            print(end="")
        try:
            tmpDict[sense+"_mean"] = tmpDF[tmpDF["Parameter"] == sense]["Value"].mean()
        except Exception:
            print(end="")
        try:
            tmpDict[sense+"_median"] = tmpDF[tmpDF["Parameter"] == sense]["Value"].median()
        except Exception:
            print(end="")
    DF = DF.append(tmpDict, ignore_index=True)  
```

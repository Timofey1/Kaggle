```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
```

```python
DATA=[]
for i in range(1,4):
    df = pd.read_csv("SOP"+str(i)+".dat", sep=" ", index_col=0, encoding="1251")
    df.columns = [str(i) for i in range(17)]
    df = df.drop("16",axis=1)
    DATA.append(df)
```

```python
DATA[0].describe().T
```

```python
plt.figure(figsize=(16,16))
sns.heatmap(DATA[0].corr())
```

```python
plt.figure(figsize=(16,7))
DATA[0].boxplot()
```

```python
plt.figure(figsize=(30, 30))

for i in DATA[0].rolling(8).mean():
    plt.subplot(16,1,int(i)+1)
    plt.title(str(int(i)+1))
    plt.plot(DATA[0].rolling(8).mean()[i], c="C"+i)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
```

```python
DATA=[]
for i in range(1,4):
    df = pd.read_csv("SOP"+str(i)+".dat", sep=" ", index_col=0, encoding="1251")
    df.columns = [str(i) for i in range(17)]
    df = df.drop("16",axis=1)
    DATA.append(df)
```

```python
DATA[0].rolling(5).mean().corr()
```

```python
correlation = DATA[0].rolling(30).mean().corr()
plt.figure(figsize=(16,16))
sns.heatmap(correlation, vmax=1, square=True,annot=True,cmap='cubehelix')
plt.title('Correlation between different fearures')
plt.show()
```

```python
correlation = DATA[1].rolling(30).mean().corr()
plt.figure(figsize=(16,16))
sns.heatmap(correlation, vmax=1, square=True,annot=True,cmap='cubehelix')
plt.title('Correlation between different fearures')
plt.show()
```

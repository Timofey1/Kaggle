```python
import tabula
import pandas as pd
import numpy as nmp
```

```python
import pdfplumber
def one_file_table(filename):
    L=[]
    pdf = pdfplumber.open(filename)
    for i in range(14):
        page = pdf.pages[i]
        try:
            table = page.extract_table()
            df = pd.DataFrame(table)
            L.append(df)
        except Exception:
            print("nope")
    df = pd.concat(L,ignore_index=True)
    df.columns = ["Раздел", "Слова", "Важность"]
    df = df.drop([0,1,2]).reset_index(drop=True)
    df["Компетенция"] = filename[:filename.find(".")]
    return df
            
```

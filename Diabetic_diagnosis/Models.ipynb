```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("clear_data.csv")
df.head().T
```

```python
models = [KNeighborsClassifier(n_neighbors=3), 
          RandomForestClassifier(random_state=42, n_estimators=100), 
          RandomForestRegressor(random_state=42,n_estimators=100),
         LogisticRegression(penalty="l1", C=10)]
```

```python
df = df.drop(["Unnamed: 0"], axis=1)
```

```python
df.readmitted.value_counts()
```

```python
X = df.drop(["readmitted"], axis=1)
y = df.readmitted
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, train_size=0.7, test_size=0.3)
```

```python
model_id = ["KNC", "RFR", "LR"]
```

```python
from sklearn.metrics import classification_report
model_ev = {}
for name,model in zip(model_id,models):
    model_ev[name] = model.fit(X_train,y_train)
    
print(model_ev)
```

```python
knc = model_ev["KNC"]
preds = knc.predict(X_test)
```

```python
for name in model_id:
    m = model_ev[name]
    print(name + " score {:.2f}".format(m.score(X_test, y_test)))
    print("_________________")
```

Тк в задании сказано что мы будем классифицировать основываясь на 10 сайтах, которые посетил человек. То мы уберем все строки где кол-во сайтов < 10. получим 22 тыс записей из 254 тыс в train_ses и 8тыс записей из 83 тыс в test.ses

```python
train_ses = train_ses[train_ses['site10'].notna()]
train_ses = train_ses.reset_index(drop=True)
```

```python
test_ses = test_ses[test_ses['site10'].notna()]
test_ses = test_ses.reset_index(drop=True)
```

C null значениями разобрались теперь обработаем datetime. Переведем дату в тип datetime чтобы было проще потом работать

```python
train_ses = train_ses.astype(td)
```
